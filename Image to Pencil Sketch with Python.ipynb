## Importing Libraries

```python
import cv2
import matplotlib.pyplot as plt
```

```python
img=cv2.imread(r'C:\Users\RAHUL KUMAR DUTTA\OneDrive\Pictures\rahul_photo .jpg')
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)

# Setting Figure Size
plt.figure(figsize=(12,5))
plt.imshow(img)

plt.title("Rahul_Kumar_Dutta's Pic")
plt.show()
```

## Converting Into Gray Scale

```python
gray_img=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
plt.figure(figsize=(12,5))

plt.imshow(gray_img,cmap='gray')
plt.title("Gray Image of Rahul Kumar Dutta")
plt.show()
```

```python
## Image Inverting
```

```python
invert_img=cv2.bitwise_not(gray_img)
plt.figure(figsize=(12,5))

plt.imshow(invert_img,cmap='gray')
plt.title('Inverted Image')
plt.show()
```

```python
## Converting into Gaussian Blur
```

```python
blur_img=cv2.GaussianBlur(invert_img,(21,21),sigmaX=0,sigmaY=0)
plt.figure(figsize=(12,5))

plt.imshow(blur_img,cmap='gray')
plt.title("Gaussian Blur Image")
plt.show()
```

## Converting Into Pencil Image

```python
pencil_img=cv2.divide(gray_img,255-blur_img,scale=255)
plt.figure(figsize=(12,5))

plt.imshow(pencil_img,cmap='gray')
plt.title("Final Pencil Sketch")
plt.show()
```
